```python
import pandas as pd
```

```python
df = pd.read_csv('evaluation_results_normal_reduced.csv')
```

```python
df
```

```python
df[df['Iteration']==0]
```

```python
df['Error Rate'] = df['Incorrect Predictions'] / (df['Correct Predictions'] + df['Incorrect Predictions'])
df
```

```python
error_by_group = df.groupby(['Category', 'Group'])['Error Rate'].mean()
print(error_by_group)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=df, x='Group', y='Error Rate', hue='Category')
plt.title('Error Rate by Group')
plt.show()
```

```python
sns.lineplot(data=df, x='Iteration', y='Error Rate', hue='Group', style='Category')
plt.title('Error Rate Over Iterations')
plt.show()
```
