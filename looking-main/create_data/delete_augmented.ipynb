```python
import os

# Caminho principal
path = r"C:\Users\madua\Documents\Mestrado\Deep Learning\Projeto Final\JAAD\images"

# Percorrer todas as subpastas e arquivos
for root, dirs, files in os.walk(path):
    for file in files:
        # Verificar se "augmented" está no nome do arquivo
        if "augmented" in file:
            file_path = os.path.join(root, file)  # Caminho completo do arquivo
            try:
                os.remove(file_path)  # Excluir o arquivo
                print(f"Arquivo excluído: {file_path}")
            except Exception as e:
                print(f"Erro ao excluir {file_path}: {e}")

print("Processo concluído!")
```
