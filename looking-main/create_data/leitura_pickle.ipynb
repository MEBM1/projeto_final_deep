```python
import pickle

# Defina o caminho para o arquivo
path = "C:\\Users\\madua\\Documents\\Mestrado\\Deep Learning\\Projeto Final\\JAAD\\data_cache\\jaad_database.pkl"

# Carrega o conteúdo do arquivo
with open(path, 'rb') as file:
    jaad_data = pickle.load(file)
```

```python
print(type(jaad_data))         # Deve ser um dicionário
print(jaad_data.keys())        # Verifique as chaves de nível superior, que geralmente são os IDs dos vídeos
```

```python
video_id = list(jaad_data.keys())[0]  # Pega o primeiro vídeo como exemplo
print(f"Dados do vídeo {video_id}:")
print(jaad_data[video_id].keys())      # Exibe as chaves internas do dicionário do vídeo (ex. 'ped_annotations')
```

```python
ped_annotations = jaad_data[video_id]["ped_annotations"]
ped_id = list(ped_annotations.keys())[0]  # Pega o primeiro pedestre como exemplo
print(f"Anotações do pedestre {ped_id}:")
print(ped_annotations[ped_id].keys())      # Exibe as chaves das anotações do pedestre (ex. 'frames', 'bbox', etc.)
```

```python
print("Attributes:", ped_annotations[ped_id].get("attributes"))
```

```python
print("Behavior:", ped_annotations[ped_id].get("behavior", "N/A"))
```
